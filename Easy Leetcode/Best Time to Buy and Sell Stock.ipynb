### Best Time to Buy and Sell Stock
input: [7,1,5,3,6,4]

output: 5

```python
class Solution:
    def maxProfit(self, prices: List[int]) -> int:
        buy, sell = 0, 0
        profit, max_profit = 0, 0
    
        for i in range(len(prices)):
            buy = prices[i]
            for j in range(i, len(prices)):
                sell = prices[j]
                profit = sell - buy
                
                if profit >= max_profit:
                    max_profit = profit
                else:
                    continue
                    
        return max_profit

prices =  [7,1,5,3,6,4]
s = Solution()
s.maxProfit(prices)
```

```python
class Solution:
    def maxProfit(self, prices: List[int]) -> int:
        buy = float("inf")
        profit = 0
        
        for price in prices:
            if price < buy:
                buy = price
            elif price - buy > profit:
                profit = price - buy
                    
        return profit

prices =  [7,1,5,3,6,4]
s = Solution()
s.maxProfit(prices)
```
